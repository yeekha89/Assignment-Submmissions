```python
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
```

```python
# Load data
digits = load_digits()
digits_data = digits.data
```

```python
# Compute WCSS
wcss_digits = []
k_range = range(1, 16)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
    kmeans.fit(digits_data)
    wcss_digits.append(kmeans.inertia_)
```

```python
# Plot Elbow Graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, wcss_digits, linestyle='-')
plt.title("The Elbow Method for Digits Dataset")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# Print conclusion
print('\nNo. The elbow method is not a great technique to yield the \
number of classes for Digits \nsince the elbow point is a bit blurry \
seemingly indicating around 8-10 classes when there \nare 10.')
```

```python
#Load libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load Iris dataset
iris = load_iris()
iris_data = iris.data
```

```python
# Compute WCSS for Iris dataset
wcss_iris = []
k_range_iris = range(1, 11)
for k in k_range_iris:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
    kmeans.fit(iris_data)
    wcss_iris.append(kmeans.inertia_)
```

```python
# Plot elbow graph
plt.figure(figsize=(8, 5))
plt.plot(k_range_iris, wcss_iris, linestyle='-')
plt.title("The Elbow Method for Iris Dataset")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# Print conclusion
print('\nYes. The elbow method yields the correct \
number of classes for the Iris dataset since the \nelbow point \
is 3 which aligns with the 3 classes for Iris.')
```
